```python
names = open("names.txt", "r").read().splitlines()
```

```python
import torch
```

```python
str = ".abcdefghijklmnopqrstuvwxyz"
stoi = {}
itos = {}
arr = list(str)
for idx, char in enumerate(arr):
    stoi[char] = idx
    itos[idx] = char
```

```python
# Create dataset
xs = []
ys = []

for name in names:
    name = ["."] + list(name) + ["."]
    for idx in range(len(name) - 1):
        xs.append(stoi[name[idx]])
        ys.append(stoi[name[idx+1]])
        
xs = torch.tensor(xs)
ys = torch.tensor(ys)
x_enc = torch.nn.functional.one_hot(xs, 27).float()
x_enc = torch.t(x_enc)
num = xs.nelement()
y_enc = torch.nn.functional.one_hot(ys, 27)
```

```python
# Creating the generator and Initializing the weights
generator = torch.Generator().manual_seed(2147483647)
weights = torch.randn(size=[27, 27], generator=generator, requires_grad=True)
```

```python
# Training Loop
learning_rate = 50
for epoch in range(100):

  # Forward Pass
  output = torch.mm(weights, x_enc)
  softmax_output = torch.exp(input=output)
  softmax_output = softmax_output / softmax_output.sum(0, keepdim=True)
  loss = -softmax_output[ys, torch.arange(num)].log().mean()
  print(loss.item())

  # Backward Pass
  weights.grad = None # Optimized way to set the initial gradient to 0
  loss.backward(retain_graph=True)

  # Weight Updates
  weights.data -= learning_rate * weights.grad
```

```python
for cnt in range(100):
    out = []
    ix = 0 # Starting with the delimiter index {".": 0}

    while True:
        x_enc = torch.nn.functional.one_hot(torch.tensor([ix]), num_classes=27).float()
        x_enc = torch.t(x_enc)
        output = torch.mm(weights, x_enc)
        probability = torch.exp(input=output)
        probability = probability / probability.sum(0, keepdim=True)

        ix = torch.multinomial(torch.t(probability), num_samples=1, replacement=True, generator=generator).item()

        # Delimiter reached again
        if ix == 0:
            break

        # Appending the current character to the output list
        out.append(itos[ix])

    print(''.join(out))
```
